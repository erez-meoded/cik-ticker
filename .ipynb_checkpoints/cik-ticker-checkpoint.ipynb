```python
import pandas as pd
```

```python
one=pd.read_csv("one.txt",sep="\t", header=0,usecols=["cik","ticker"])
two=pd.read_csv("two.txt",sep="\t",header=0,usecols=["cik","ticker"])
sub=pd.read_csv("sub.txt", sep='\t', header=0,usecols=["cik"])
```

```python
three=two.merge(one,"outer",["cik","ticker"])
```

```python
four=three.merge(sub,"outer",["cik"])
```
