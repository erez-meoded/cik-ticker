```python
import pandas as pd
```

```python
import requests, json
import time
import datetime
#concat companies dataframe and save it to csv
try:
    companies
except NameError:
    companies = pd.DataFrame()

for i in range(200):#change it to 200 if you want to fetch ALL companies in EDGAR public data
    r = requests.get('http://edgaronline.api.mashery.com/v2/companies.json?filter=cik btw "000'+ str(i).zfill(3) + '0000" and "000'+ str(i).zfill(3) +'9999" &limit=1000&appkey=xxxxxxxxxxxxxxx')
    if r.status_code == 200:#OK
        json_data = json.loads(r.text)
        print(i,len(json_data['result']['rows']))
        for row in json_data['result']['rows']:
            companies = pd.concat([companies, pd.DataFrame(row['values']).set_index('field').transpose()], ignore_index=True).fillna('').sort_values('cik')
    else:
        print(r.text) #print error/info
    time.sleep(0.5) #limit of 2 calls per second
companies.to_csv('companies-'+ str(datetime.datetime.now().isoformat()).replace(':','') + '.csv',',',index=False)
companies
```
