```python
import pandas as pd
import numpy as np
```

```python
one=pd.read_csv("one.txt",sep="\t", header=0,usecols=["cik","ticker"], dtype={'cik': np.int64})
two=pd.read_csv("two.txt",sep="\t",header=0,usecols=["cik","ticker"], dtype={'cik': np.int64})
sub=pd.read_csv("sub.txt", sep='\t', header=0,usecols=["cik"], dtype={'cik': np.int64})
```

```python
three=two.merge(one,"outer",["cik","ticker"])
```

```python
four=three.merge(sub,"outer",["cik"])
```
